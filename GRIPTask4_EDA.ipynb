<b> From the above graph, following conclusions are made:

<b>3.1.4: Category.</b>

<h2><b><span style='color:MediumVioletRed'>The above result depicts that,

    <li>If Discount is between 0% to 20%, Sales is average, but, the Superstore gains Profit.</li>
    </ol></span></b></h2>

```python
temp_data = data.loc[(data['Segment'] == 'Corporate') & (data['Discount'] == 0.6)]
temp_data['Profit'].plot.hist(bins=20)
```

```python
temp_data = data.loc[(data['Segment'] == 'Home Office') & (data['Discount'] == 0.2)]
temp_data['Profit'].plot.hist(bins=20)
```

<b> For all the Segments, If the Super Store offers Discount less than 30%, it gains some Profit. 
But, if the Discount is more than 50%, it faces loss. </b>

<b>3.3.6: Analysis based on Grouping - Analyse Sales, Profit based on Discount for Sub-Category wise.</b>

```python
data_sub = data.groupby(['Sub-Category'])[['Discount', 'Sales', 'Profit']].mean()
print(data_sub)
```

<b>Viewing the above result, visually using Pie chart.</b>

```python
plt.figure(figsize=(14,14))
plt.pie(data_sub['Sales'], labels=data_sub.index, autopct='%1.1f%%')
plt.title('Sales Analysis by Sub-Category wise', fontsize=15)
plt.legend()
plt.xticks(rotation=90)
```

<b>Profit analysis Sub-Categorywise.</b>

```python
plt.figure()
sc = data.groupby(['Sub-Category'])[['Profit']].sum().plot.bar(color='DarkSlateGrey', figsize=(10,6))
plt.title("Profit Analysis Sub-Categorywise", fontsize=15, color='green')
plt.ylabel("Profit", fontsize=15, color='red')
plt.xlabel("Sub-Categories", fontsize=15, color='red')
plt.show()
```

<b><span style = 'color:MediumVioletRed'><span style = 'font-size:20px'>Copiers sub-category has high Sales and yields a good Profit.</span></span> </b>

<b>3.3.7: Statewise Sales, Profit Analysis.</b>

```python
sw = data['State'].value_counts().nlargest(10)
print(sw)
```

<b> Visualizing the above Result.</b>

```python
data_state = (data.groupby(['State'])['Sales', 'Profit'].agg('sum')).plot.bar(figsize=(20,8), width=0.8, color=['blue', 'red'])
plt.xlabel("States in U.S.", fontsize=15)
plt.title("Statewise Sales, Profit.", fontsize=20)
```

<h2><b><span style= 'color:MediumVioletRed'>The Conclusion Arrived is:<br>The Super store in California State has 
        high Sales & Profit. Next comes New York and Washington. In Texas also the Sales is high.</span></b></h2>

<b>Pie-Chart.</b>

```python
tempdata = data.groupby(['State'])[['Sales', 'Profit']].count()
plt.figure(figsize=(27,27))
plt.pie(tempdata['Sales'], labels=tempdata.index, autopct='%1.1f%%')
plt.title("Sales & Profit Analysis by State wise.",fontsize=20)
plt.legend()
plt.xticks(rotation=90)
```

<b><span style = 'color:green'><span style= 'font-size:18px'>The above Pie chart clearly shows that California State has high Sales and produces High Profit. 
<br>The weak areas are:Maine, Iowa, Idaho, North Dakota, Vermont, Montana, New Hampshire, South Dakota.</span></span></b>

<b>3.3.8: Analysis of Sales, Profit by Regionwise, Segmentwise & Categorywise.</b>

```python
data_reg = (data.groupby(['Region', 'Category', 'Segment'])['Sales', 'Profit'].sum())/1000
print(data_reg)
```

<b> Visualizing the above Result.</b>

```python
data_reg.plot.bar(figsize=(20,10), width=1, color=['green','red'])
plt.title("Sales, Profit  by Regionwise, Segmentwise & Categorywise.", fontsize=20)
plt.xlabel("Region, Category, Segment", fontsize=15)
plt.ylabel("Sales & Profit", fontsize=15)
plt.show()
```

<h3><b><span style='color: MediumVioletRed'>The Following Conclusions arrived from the above graph.<br>
<ul><li>The 'West' Region, 'Technology' Category & 'Consumer' Segment have high Sales and produces high Profit.</li>
    <li>The weak areas to be concentrated are: <i> South Region, Furniture & Office Supply Categories & Home Office 
        Segment.</i></li>
    </ul></b></span></h3>
